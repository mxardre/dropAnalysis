```python
import numpy 
import math
import matplotlib.pyplot as plt
%matplotlib inline
path='/Users/maxime/Documents/experiences/distribution/bioconvection/BioconvectionVideo_thermalise/'
a = genfromtxt(path+'analysisPIV2017-07-20_therm.csv', delimiter=',')
v=a[:,6]
vv = [x for x in v[1:] if not x==-1]
vv = [x/(4*10)*60 for x in vv ]


s=numpy.std(vv);
moy=numpy.mean(vv);
N=len(vv);
bin=3.5*s/(N**(1/3));
nbin=numpy.linspace(min(vv),max(vv), (max(vv)-min(vv))/bin);


plt.hist(vv, nbin, alpha=0.5)
#plt.legend(loc='upper right')
plt.title('speed of plumes %.d µm/min' % moy )
plt.xlabel("µm/min")
plt.ylabel('count')
#plt.ylim([0, 50])
plt.savefig(path+'histVitesse.png')
plt.show()

```
