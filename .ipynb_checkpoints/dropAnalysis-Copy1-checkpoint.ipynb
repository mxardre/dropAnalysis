```python
#import the droplet file and load them in a dataframe
import pandas as pd
import numpy as np
from os import listdir
from os.path import isfile, join

dt = np.dtype([('time','<i4'),]+[(x, '<i2') for x in ['pmt1','pmt2','pmt3']])

path='/Users/maxime/Documents/experiences/milli/MILLIDROP_manip/2017-07-28_wt_dilutionKITP/'
drpfiles=[path+'droplets/' + f for f in listdir(path + 'droplets/') if isfile(join(path + 'droplets/',f))]

i=0
df={}
for file in drpfiles:
    df[i]=pd.read_csv(file)
    i+=1

```

```python
#plot all the data roughly
import matplotlib.pyplot as plt

for i in range(0,len(df)):
    p=df[i]
    plt.scatter(p.time/3600, p.fluo_2_mean)

plt.show()
```

```python
#create the template file in hiccup mode
tpfile=pd.read_csv(path+'template.csv', header=4)
tpFileOrd=tpfile.set_index('order')

dropMap=[]

for i in range(0,len(tpFileOrd)-1,2):
    for j in range(tpFileOrd.droplet_number[i]):
        if j%2==0 :
            #dropMap.append([tpFileOrd.well[i], tpFileOrd.description[i]])
            dropMap.append([tpFileOrd.well[i], tpFileOrd.description[i]])
        else : 
            #dropMap.append([tpFileOrd.well[i+1], tpFileOrd.description[i+1]])
            dropMap.append([tpFileOrd.well[i+1], tpFileOrd.description[i+1]])

dropMap=np.array(dropMap)
#dropMap[:,1]
#dropMap = np.delete(dropMap, (0), axis=0)
```

```python
#label='PvdS-25'
#label='PvdS-5'
#label='Empty'
#label='KB'
#label='SBW25-25'

def plotLabel(df, dropMap, label):
    j=0
    for i,content in enumerate(dropMap[:,1]):
        if content==label:
            p=df[i]
            plt.scatter(p.time/3600, p.fluo_2_mean)
            j+=1
    plt.ylim([0, 5])
    plt.xlim([0, 45])
    plt.show()

#plotLabel(df, dropMap, 'SBW25-25')
plotLabel(df, dropMap, 'Separator')
plotLabel(df, dropMap, '100')
```

```python
def plotLabelSemiLog(df, dropMap, label):
    j=0
    for i,content in enumerate(dropMap[:,1]):
        if content==label:
            p=df[i]
            plt.scatter(p.time/3600, np.log(p.fluo_3_mean))
            j+=1
    plt.ylim([0, np.log(5)])
    plt.xlim([0, 25])
    plt.show()
    

#plotLabelSemiLog(df, dropMap, 'PvdS-5')
#plotLabelSemiLog(df, dropMap, 'PvdS-25')
```
